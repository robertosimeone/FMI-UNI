```python
import sys
import numpy as np
import sklearn
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%matplotlib inline
```

```python
# Kmeans
normal = pd.read_csv("normal.csv")
unbalanced = pd.read_csv("unbalance.csv")
normal_coordinates = pd.read_csv("./remaked_info/normal_coordinates.csv")
unbalanced_coordinates = pd.read_csv("./remaked_info/un_coordinates.csv")
normal_results = pd.concat([normal_coordinates,normal],axis=1)
unabalanced_results = pd.concat([unbalanced_coordinates,unbalanced],axis=1)

# Kmeans++
normal_plus = pd.read_csv("normal_plus_plus.csv")
unbalanced_plus = pd.read_csv("unbalance_plus_plus.csv")



normal_plus_results = pd.concat([normal_coordinates,normal_plus],axis=1)
unabalanced_plus_results = pd.concat([unbalanced_coordinates,unbalanced_plus],axis=1)
# plt.figure(figsize=(10,10))
# plt.scatter(normal_results.iloc[:,0], normal_results.iloc[:,1], color = 'violet')
# plt.scatter(a, c, color = 'red')
# plt.grid(True)
# plt.show()
# plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the number of colors
num_colors = 20

# Generate the color list
colors = []
for i in range(0, num_colors):
    # Generate a random RGB color
    rgb = np.random.randint(0, 255, 3)
    color = '#%02x%02x%02x' % (rgb[0], rgb[1], rgb[2])
    colors.append(color)
```

```python
plt.figure(figsize=(8,8))
plt.grid(True)
colors = ['green','red','blue','orange','pink']
uniq_values = normal_results.iloc[:,2].unique()
print(uniq_values) 
for index,el in enumerate(uniq_values):
    current = normal_results.loc[normal_results["cluster"] == el]
    plt.scatter(current.loc[:,'x'], current.loc[:,'y'], color = colors[index])
plt.title("Standard Kmeans - normal data")
plt.show()    
```

```python
plt.figure(figsize=(8,8))
plt.grid(True)
colors = ['green','red','blue','yellow','pink','orange','black','indigo']
uniq_values = unabalanced_results.iloc[:,2].unique()
print(uniq_values)
for index,el in enumerate(uniq_values):
    current = unabalanced_results.loc[unabalanced_results["cluster"] == el]
    plt.scatter(current.loc[:,'x'], current.loc[:,'y'], color = colors[index])
plt.title("Standard Kmeans - unbalanced data")
plt.show()    
```

```python
plt.figure(figsize=(8,8))
plt.grid(True)
colors = ['green','red','blue']
uniq_values = normal_plus_results.iloc[:,2].unique()
print(uniq_values) 
for index,el in enumerate(uniq_values):
    current = normal_plus_results.loc[normal_plus_results["cluster"] == el]
    plt.scatter(current.loc[:,'x'], current.loc[:,'y'], color = colors[index])
plt.title("Kmeans++ - normal data")
plt.show()  
```

```python
plt.figure(figsize=(8,8))
plt.grid(True)
colors = ['green','red','blue','yellow','pink','orange','black','indigo']
uniq_values = unabalanced_plus_results.iloc[:,2].unique()
print(uniq_values)
for index,el in enumerate(uniq_values):
    current = unabalanced_plus_results.loc[unabalanced_plus_results["cluster"] == el]
    plt.scatter(current.loc[:,'x'], current.loc[:,'y'], color = colors[index])
plt.title("Kmeans++ - unbalanced data")
plt.show() 
```
